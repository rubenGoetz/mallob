```python
import re
import os

def get_runtime(log):
    runtime = 0.
    # Mallob style output
    s = re.search("RESPONSE_TIME #[0-9]+ ([0-9]+.[0-9]+)", log)
    if s:
        runtime = float(s.groups()[0])
        return runtime
    # Gimsatul style output
    s = re.search("c wall-clock-time:( )+([0-9]+.[0-9]+)", log)
    if s:
        runtime = float(s.groups()[1])
        return runtime
    # Kissat style output
    s = re.search("c process-time:( )+([0-9]+m [0-9]+s)( )+([0-9]+.[0-9]+) seconds", log)
    if s:
        runtime = float(s.groups()[3])
        return runtime
    return 0

log = open('../logs_serial/hash_table_find_safety_size_23.cnf/out_file').read()
get_runtime(log)
```
